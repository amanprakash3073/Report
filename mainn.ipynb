```python
import pandas as pd
from scipy.stats import pearsonr, spearmanr  # <-- make sure this line is included

# Mean Magnitude and Depth data
data = {
    'Country': ['Canada', 'Chile', 'India', 'Myanmar', 'Mexico', 'Nepal', 'Pakistan', 'Peru', 'Turkey'],
    'Magnitude': [2.064279359430605, 5.360888691513608, 5.328085867620751, 4.791699218750001,
                  4.1492309962010845, 4.474325153374234, 4.077258566978193, 4.538738501971091, 4.388630865989356],
    'Depth': [14.69564056939502, 59.62202708137819, 28.36314847942755, 42.3600419921875,
              39.097063922393126, 13.031924846625767, 22.185150571131878, 75.73780113885239, 14.462474600870825]
}

df = pd.DataFrame(data)

# Pearson correlation (linear)
pearson_corr, pearson_p = pearsonr(df['Magnitude'], df['Depth'])

# Spearman correlation (rank-based)
spearman_corr, spearman_p = spearmanr(df['Magnitude'], df['Depth'])

print("=== Correlation between Mean Magnitude and Depth ===")
print(f"Pearson correlation: {pearson_corr:.3f} (p = {pearson_p:.4f})")
print(f"Spearman correlation: {spearman_corr:.3f} (p = {spearman_p:.4f})")
```

```python
q1=df['magnitude'].quantile(0.25)
q2=df['magnitude'].quantile(0.50)
q3=df['magnitude'].quantile(0.75)
```

```python
def classify(val):
    if val<=q1:
        return 'q1'
    elif val<=q2:
        return 'q2'
    elif val<=q3:
        return 'q3'
    else:
        return 'q4'
```

```python
df['quartile_cate']=df['magnitude'].apply(classify)
df['rank']=df['magnitude'].astype(int)
print(df)
```

```python
data={
    'country':['canada','chile','india','maynmar','maxico','nepal','pakistan','peru','turkey'],
    'magnitude' :[2.06,5.36,5.33,4.76,4.15,4.47,4.08,4.54,4.39],
    'depth' :[14.69,59.62,28.36,42.36,39.10,13.03,22.18,75.74,14.46]
}
df=pd.DataFrame(data)
```

```python
countries=df['country'].unique()
store=[]
for country in countries:
    temp=df[df['country']==country]
    if(len(temp))>1:
        r_mag_depth=pearsonr(temp['magnitude'],temp['depth'])
        store.append([country,r_mag_depth])
    else:
        store.append([country,None,None])
result=pd.DataFrame(store,columns=['country','correlation','p_val'])
print(result)
```
