```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
analysis = pd.DataFrame({
    'country': ['canada','chile','india','myanmar','mexico','nepal','pakistan','peru','turkey'],
    'magnitude': [2.06,5.36,5.33,4.76,4.15,4.47,4.08,4.54,4.39],
    'depth': [14.69,59.62,28.36,42.36,39.10,13.03,22.18,75.74,14.46],
    'freq': [2248,59672,2273,1024,27113,652,963,11415,11376],
    'gdp_billion_usd': [2200,352,3400,65,1800,40,375,250,900]
})
```

```python
plt.boxplot(analysis['magnitude'], patch_artist=True)
plt.title('Distribution of Earthquake Intensity')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
analysis_sorted = analysis.sort_values(by='magnitude', ascending=False)
plt.bar(analysis_sorted['country'], analysis_sorted['magnitude'], color='coral')
plt.title('Average Earthquake Magnitude by Country')
plt.ylabel('Magnitude')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.show()
```

```python
global_avg = analysis['magnitude'].mean()
analysis = analysis.sort_values(by='magnitude', ascending=False)
plt.figure(figsize=(10,6))
plt.bar(analysis['country'], analysis['magnitude'], color='skyblue', label='Country Magnitude')
plt.axhline(y=global_avg, color='red', linestyle='--', linewidth=2, label=f'Global Avg = {global_avg:.2f}')

plt.title('Comparison of Earthquake Intensity with Global Average')
plt.xlabel('Country')
plt.ylabel('Average Magnitude')
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

analysis = pd.DataFrame({
    'country':['Canada','Chile','India','Myanmar','Mexico','Nepal','Pakistan','Peru','Turkey'],
    'magnitude':[2.06,5.36,5.33,4.76,4.15,4.47,4.08,4.54,4.39]
})

mean_mag = analysis['magnitude'].mean()
std_mag = analysis['magnitude'].std()

upper_limit = mean_mag + std_mag
lower_limit = mean_mag - std_mag

plt.figure(figsize=(10,6))
plt.bar(analysis['country'], analysis['magnitude'], color='lightseagreen', label='Country Magnitude')
plt.axhline(y=mean_mag, color='blue', linestyle='--', label=f'Mean = {mean_mag:.2f}')
plt.axhline(y=upper_limit, color='red', linestyle='--', label=f'+1 SD = {upper_limit:.2f}')
plt.axhline(y=lower_limit, color='orange', linestyle='--', label=f'-1 SD = {lower_limit:.2f}')
plt.title("Earthquake Magnitude Distribution with Mean and Standard Deviation")
plt.xlabel("Country")
plt.ylabel("Average Magnitude")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

print(f"Mean Magnitude: {mean_mag:.2f}")
print(f"Standard Deviation: {std_mag:.2f}")
print(f"Upper Limit (+1 SD): {upper_limit:.2f}")
print(f"Lower Limit (-1 SD): {lower_limit:.2f}")
```
