```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'country':['canada','chile','india','maynmar','maxico','nepal','pakistan','peru','turkey'],
    'magnitude' :[2.06,5.36,5.33,4.76,4.15,4.47,4.08,4.54,4.39],
    'depth' :[14.69,59.62,28.36,42.36,39.10,13.03,22.18,75.74,14.46],
    'freq' :[2248,59672,2273,1024,27113,652,963,11415,11376]
}

df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='country', y='freq', data=df, palette='Blues_d')
plt.title("Earthquake Frequency  Country")
plt.xlabel("Country")
plt.ylabel("No. of Earthquakes")
plt.xticks(rotation=25)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='depth', y='magnitude', hue='country', s=100, data=df)
plt.title("Depth vs Magnitude of Earthquakes")
plt.xlabel("Depth (km)")
plt.ylabel("Magnitude")
plt.show()
```

```python
plt.figure(figsize=(12,5))

# Boxplot for Magnitude (Horizontal)
plt.subplot(1,2,1)
sns.boxplot(x=df['magnitude'], color='orange')
plt.title("Outlier Detection in Magnitude")
```

```python
plt.subplot(1,2,2)
sns.boxplot(x=df['depth'], color='lightblue')
plt.title("Outlier Detection in Depth")
```

```python
correlation = df[['magnitude', 'depth', 'freq']].corr()
print("correlation  is :",correlation)
```

```python
df['risk_score'] = df['magnitude'] * df['freq']
ranked = df.sort_values('risk_score', ascending=False)
print(ranked[['country','risk_score']])
```
